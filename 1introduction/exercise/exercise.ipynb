```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('height.vs.temperature.csv')
data.head(15)
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(data['height'], data['temperature'], c='red')
font = {'family': 'serif',
        'color':  'white',
        'weight': 'normal',
        'size': 20,
        }
plt.xlabel('height', fontdict=font)
plt.ylabel('temp', fontdict=font) 
plt.show()
```

```python
x = data['height'].values.reshape(-1, 1)
y = data['temperature'].values.reshape(-1, 1)
reg = LinearRegression()
reg.fit(x, y)
```

```python
a = reg.coef_[0][0]
b = reg.intercept_[0]
```

```python
a,b
```

```python
print('relationship betweem height and weight is y = %s *x + %s' % (a, b))
```

```python
predicts = reg.predict(x)
```

```python
predicts
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(data['height'], data['temperature'], c='red')
plt.plot(data['height'], predicts, c='blue', linewidth=5)
font = {'family': 'serif',
        'color':  'white',
        'weight': 'normal',
        'size': 20,
        }
plt.xlabel('height', fontdict=font)
plt.ylabel('temp', fontdict=font) 
plt.show()
```
